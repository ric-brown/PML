### Revisiting the Problem Statement:
![A clean image, the corrupted image, and a pairwise MRF](./figures/problem_statement_diagram.png)

The goal of this problem is to recover the clean image (a) from a noisy input (b). Of course, we cannot recover (a) exactly since information has been lost in the noise. The graphical model we use is a pairwise MRF as shown in (c). A pixel has four neighbors. We are going to use Markov Random Fields (MRFs) to model the distribution of natural images and restore the clean image given an noisy input image. Observed image $y_i \in \{-1,+1\}$, and $i = 1,\cdots,D$ indexes pixels in the lattice. The
original noise free image is $x_i \in \{-1, +1\}$. The noisy image $y_i$ is obtained by randomly flipping the sign of pixels with some probability (10% of pixels, in this case).

There are two types of cliques in this MRF. For $\{x_i,y_i\}$, we define the energy function as $-\eta x_i y_i\ (\eta > 0)$.  Lower energy is achieved when $x_i$ and $y_i$ have the same sign and a higher energy when they have the opposite sign. For a pair of variables $\{x_i, x_j\}$ where $i$ and $j$ are indices of neighboring pixels, we want the energy to be lower when they have the same sign than when they have opposite sign. So the energy function is $-\beta x_i j_i\ (\beta > 0)$. Lastly, we have an energy term $hx_i$ for each pixel $i$ to bias the model towards one particular sign (either $+$ or $−$).

The final energy function for the model takes the form

$E(\vec{x}\vec{y}) = h\displaystyle\sum\limits_i x_i - \beta \displaystyle\sum\limits_{\{i,j\}} x_ix_j - \eta\displaystyle\sum\limits_i x_iy_i$

which defines a joint distribution over $\vec{x}$ and $\vec{y}$ given by

$p(\vec{x},\vec{y}) = \dfrac{1}{\vec{\mathcal{Z}}}exp\{-E(\vec{x}\vec{y})\}$

Now implement Coordinate-descent algorithm as below on this:
1. Initialize $\{x_i\} (x_i = y_i)$
2. Loop over $\{x_i\}$. For each $x_i$, fix the neighborhood and see whether $−x_i$ would
decrease the energy. If so, then flip $x_i$; otherwise, continue.
3. Stop when no changes can be made for $x$.

Now make some initial guess for the parameters $h$, $\beta$, $\eta$ so that the above algorithm converges and then adjust them until you can get up to 96% recovery or better. Record your parameter values and the number of iterations when you acheived this recovery.

## MRF

```python
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
## Implement your coordinate-descent algorithm here
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

# MRF Energy Function
def compute_energy(x, y, h, beta, eta):
    """
    E(x,y) = h*sum(x_i) - beta*sum(x_i*x_j) - eta*sum(x_i*y_i)
    """
    bias_term     = h * np.sum(x)
    pairwise_term = beta * (
        np.sum(x[:-1, :] * x[1:, :]) +   # vertical neighbors
        np.sum(x[:, :-1] * x[:, 1:])     # horizontal neighbors
    )
    obs_term = eta * np.sum(x * y)
    return bias_term - pairwise_term - obs_term
```

```python
# Parameters
h          = 0.0
beta       = 2.0
eta        = 1.5
max_sweeps = 100

x = copy.copy(y)
```

```python
# ── Coordinate Descent ───────────────────────────────────────
sweep_log = []

print(f"Starting coordinate descent | h={h}, beta={beta}, eta={eta}")
print(f"{'Sweep':>6}  {'Flips':>8}  {'Recovery %':>12}  {'Energy':>14}")
print("-" * 50)

for sweep in range(max_sweeps):
    n_flips = 0

    for i in range(H):
        for j in range(W):

            # Sum in-bounds neighbors
            neighbor_sum = 0.0
            if i > 0:    neighbor_sum += x[i-1, j]
            if i < H-1:  neighbor_sum += x[i+1, j]
            if j > 0:    neighbor_sum += x[i, j-1]
            if j < W-1:  neighbor_sum += x[i, j+1]

            # Flip x_i if it decreases energy
            # ΔE < 0 when x_i * (h - β*Σneighbors - η*y_i) > 0
            local_field = h - beta * neighbor_sum - eta * y[i, j]
            if x[i, j] * local_field > 0:
                x[i, j] = -x[i, j]
                n_flips += 1

    energy   = compute_energy(x, y, h, beta, eta)
    recovery = np.mean(x == x_clean) * 100
    sweep_log.append((sweep + 1, n_flips, recovery, energy))
    print(f"{sweep+1:>6}  {n_flips:>8}  {recovery:>11.2f}%  {energy:>14.1f}")

    if n_flips == 0:
        print(f"\nConverged after {sweep+1} sweep(s).")
        break

print(f"\nFinal recovery: {recovery:.2f}%")
```

Below is a function to optimize a range of features and find the best preforming features. A numpy array was added to the pixel loop computing the neighbor's sum in a numpy array so it is computationally efficent an my computer doesn't melt.

```python
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
## Simulated Annealing + Coordinate Descent
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

def simulated_annealing(y, x_clean, h, beta, eta,
                        T_start=2.0, T_end=0.01, max_sweeps=50):
    x      = copy.copy(y)
    H, W   = x.shape
    sweep_log = []

    for sweep in range(max_sweeps):
        # Temperature schedule: exponential cooling
        T = T_start * (T_end / T_start) ** (sweep / max_sweeps)

        n_flips  = 0
        n_accept = 0

        for i in range(H):
            for j in range(W):
                neighbor_sum = 0.0
                if i > 0:    neighbor_sum += x[i-1, j]
                if i < H-1:  neighbor_sum += x[i+1, j]
                if j > 0:    neighbor_sum += x[i, j-1]
                if j < W-1:  neighbor_sum += x[i, j+1]

                local_field = h - beta * neighbor_sum - eta * y[i, j]
                delta_E     = -2 * x[i, j] * local_field

                if delta_E < 0:
                    # Always accept energy-decreasing flips
                    x[i, j]  = -x[i, j]
                    n_flips  += 1
                    n_accept += 1
                elif np.random.rand() < np.exp(-delta_E / T):
                    # Occasionally accept energy-increasing flips
                    x[i, j]  = -x[i, j]
                    n_accept += 1

        energy   = compute_energy(x, y, h, beta, eta)
        recovery = np.mean(x == x_clean) * 100
        sweep_log.append((sweep + 1, T, n_flips, n_accept, recovery, energy))
        print(f"Sweep {sweep+1:>3} | T={T:.4f} | "
              f"flips={n_flips:>6} | accepted={n_accept:>6} | "
              f"recovery={recovery:.2f}% | energy={energy:.1f}")

        # Stop early if fully converged at low temperature
        if T < 0.05 and n_flips == 0:
            print(f"\nConverged at sweep {sweep+1}")
            break

    return x, sweep_log, recovery
```

```python
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
## Parameter Search with Simulated Annealing
## +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

best_recovery = 0
best_params   = None
best_x        = None
best_log      = None
results       = []

for h in [0.0, -0.1]:
    for beta in [1.5, 2.0, 2.5]:
        for eta in [1.0, 1.5, 2.0]:
            for T_start in [1.0, 2.0, 5.0]:

                print(f"\nh={h}, beta={beta}, eta={eta}, T_start={T_start}")
                print("-" * 70)

                x_trial, log, recovery = simulated_annealing(
                    y, x_clean,
                    h=h, beta=beta, eta=eta,
                    T_start=T_start, T_end=0.01, max_sweeps=50
                )

                results.append((recovery, h, beta, eta, T_start))

                if recovery > best_recovery:
                    best_recovery = recovery
                    best_params   = (h, beta, eta, T_start)
                    best_x        = copy.copy(x_trial)
                    best_log      = log
                    
```

```python
# Top 5
results.sort(reverse=True)
print(f"\n── Top 5 Parameter Combinations ────────────────────")
for r, h, b, e, t in results[:5]:
    print(f"  h={h}, beta={b}, eta={e}, T_start={t} → {r:.2f}%")

print(f"\nBest: h={best_params[0]}, beta={best_params[1]}, "
      f"eta={best_params[2]}, T_start={best_params[3]} → {best_recovery:.2f}%")

# Set best result for Cell 4
x = best_x
sweep_log = best_log
recovery = best_recovery
h, beta, eta, T_start = best_params
```
